```python
# Daily Aggregation by year_month
df_cleansed['creation_date'] = pd.to_datetime(df_cleansed['creation_date'])
df_cleansed['year_month_day'] = df_cleansed['creation_date'].dt.to_period('D') # daily
techiness = df_cleansed.groupby(['year_month_day'])['techiness'].mean().reset_index()
```
