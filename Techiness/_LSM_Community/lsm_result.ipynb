# LSM_new (Daily Agg. vr.) Experiment Result

```python
# Load Pickle
import pickle
import pandas as pd
import numpy as np

with open(file='LSM_new_community_result.pickle', mode='rb') as f:
    result=pickle.load(f)
result
```

```python
# Preprocess
average_lsm = result.groupby(['year_month', 'community'])['lsm_score'].mean().reset_index()
average_lsm['year_month_day'] = average_lsm['year_month'].astype(str)
average_lsm = average_lsm[['year_month_day', 'community', 'lsm_score']]
average_lsm
```

### Merge to df_final_pre6.csv

```python
# average_lsm = year_month, community, lsm_score
## add to the data below.
df_final_pre6 = pd.read_csv("/data1/StackOverflow/_Robustness/Techiness/df_final_pre6.csv")
df_final_pre6
```

```python
df_merge = pd.merge(df_final_pre6, average_lsm, on = ['year_month_day', 'community'], how = 'left')
df_merge = df_merge.rename(columns = {'lsm_score':'lsm'})
df_merge['ln_lsm'] = np.log(df_merge['lsm'])
df_merge
```

```python
df_merge.to_csv('df_final_pre7.csv', index = False)
```
